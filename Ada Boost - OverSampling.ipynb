# OVER SAMPLING

```python
Y.value_counts().plot(kind = 'bar')
```

```python
from imblearn.over_sampling import RandomOverSampler, SMOTE
```

```python
ros = RandomOverSampler()

ros
```

```python
X.shape, Y.shape
```

```python
X_ros, Y_ros = ros.fit_resample(X, Y)
```

```python
X_ros.shape, Y_ros.shape
```

```python
Y_ros.value_counts()
```

```python
Y_ros.value_counts().plot(kind = 'bar')
```

```python
smote = SMOTE()
```

```python
Xsmote, Ysmote = smote.fit_resample(X, Y)
```

```python
Xsmote.shape, Ysmote.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
Xtrain, Xtest, Ytrain, Ytest = train_test_split(Xsmote, Ysmote, test_size = 0.3)
```

```python
Xtrain.shape, Xtest.shape, X.shape
```

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
ada = AdaBoostClassifier()
```

```python
ada.fit(Xtrain, Ytrain)
```

```python
ada.score(Xtest, Ytest)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
```

```python
adarfc = AdaBoostClassifier(base_estimator = rfc, algorithm = 'SAMME')
```

```python
adarfc.fit(Xtrain, Ytrain)
```

```python
adarfc.score(Xtest, Ytest)
```
